```python
size=df.shape
```

```python
size[0]
```

```python
s=[[]*size[1]]*size[0]
index=0
for col in df.columns: 
    
    s[index]=df[col].values.tolist()
    index+=1
```

```python
print(len(s))
parentNodeName='Person'
xmlChildElments=''
for i in range(size[0]):
    xmlChildElments+='<'+parentNodeName+'>'
    for j in range(size[1]):
        xmlChildElments+='<'+df.columns[j]+'>'+str(s[j][i])+'</'+df.columns[j]+'>\n'
    xmlChildElments+='</'+parentNodeName+'>'
```

```python
#print(s[])
print(xmlChildElments)
```

```python
fs=open('./patientXML.xml','w')
fs.write(xmlChildElments)
```
